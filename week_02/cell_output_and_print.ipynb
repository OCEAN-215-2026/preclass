# Remark: cell output and the `print()` function

As you have already seen, a python cell generally prints out the result of its last line, unless the result of that line is `None` or if the last line is an assignment operation:

```python
# the last line is printed out
x = 5
y = 3
x + y
```

```python
# the last line is an assignment operation - there is no output
x = 5
y = 3
z = x + y
```

```python
# The last line is None - no output either
x = None
x
```

If you want to print out intermediate results in the middle of a code cell, you'll need to use the `print()` function:

```python
x = 5
y = 3
z = x + y
print(z)
w = z ** 2
w
```

Another use of the print function is to print a concatenation of values, using the fact that the `print()` function can accept an arbitrary number of arguments. For example:

```python
x = 3
print("The value of x is", x)
```

Here we passed two values to the `print()` function, the first being the string `"The value of x is"` and the second being the numerical variable `x` (which evaluates to `3` when the print function executes). Notice that a space is automatically inserted between the two arguments.

An alternative way to achieve the above output is:

```python
x = 3
print("The value of x is" + " " + str(x))
```

In this case there is only **one** argument to the print function, which is the result of the operation `"The value of x is" + " " + str(x)`. Note that I have to insert the extra space by hand (and could have simplified the argument to `"The value of x is " + str(x)`), and explicitly instruct python to convert the numeric value `3` to a string so that the `+` would work.

Finally, it should be noted that cell output and the result of `print()` function can have different formatting. One example involves the numpy array, which we'll cover in much greater details in week 4. For now, just note that the second line below defines a numpy array object, whose printout looks different from what the cell output.

```python
import numpy as np
x = np.array([1, 2, 3])
print(x)
x
```
